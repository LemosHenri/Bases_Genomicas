```python
import pandas as pd
import numpy as np
import os
```

```python
# Leitura do dataframe contendo os dados brutos
ORGANISMO = 'Aedes'
df = pd.read_csv(f'../Data/{ORGANISMO}/dados_brutos.csv')
df.head()
```

```python
# Verificando colunas do dataframe
print(df.columns)
```

```python
# Filtrando os dados relevantes
main_df = df[['Run_acc', 
            'Biosample',
            'Bioproject',
            'Organism_taxid',
            'Platform_instrument_model', 
            'Statistics_total_spots', 
            'Statistics_total_bases', 
            'Statistics_total_size',
            'Organism_ScientificName', 
            'LIBRARY_STRATEGY',
            'LIBRARY_SOURCE', 
            'LIBRARY_SELECTION', 
            'LIBRARY_CONSTRUCTION_PROTOCOL', 
            'createdate',
]]
main_df.columns = ['SRA acc', 
                'Biosample acc', 
                'Bioproject acc', 
                'Taxonomia acc', 
                'Instrumento', 
                'Amostras', 
                'Bases', 
                'Tamanho',
                'Organismo',
                'Estratégia',
                'Recurso',
                'Seleção',
                'Protocolo',
                'Criação',
]

main_df.head()
```

```python
# Ajustando tipos não recomendados
main_df['Tamanho'] = main_df['Tamanho'].astype(np.float32)
main_df['Bases'] = main_df['Bases'].astype(np.float32)
main_df['Taxonomia acc'] = main_df['Taxonomia acc'].astype(str)

# Avaliando strings mal formatadas
main_df['Recurso'] = main_df['Recurso'].str.title()
main_df['Seleção'] = main_df['Seleção'].str.title()

# Ajustando dados numéricos
main_df['Tamanho'] = main_df['Tamanho'].apply(lambda x: round(x / 1e6, 2))
main_df['Bases'] = main_df['Bases'].apply(lambda x: round(x / 1e9, 1))
```

```python
# Verificando novo estado
main_df.head()
```

```python
diretorio = f'../Data/{ORGANISMO}'

os.makedirs(f'{diretorio}', exist_ok = True)
main_df.to_csv(f'{diretorio}/dados_tratados_{ORGANISMO}.csv', index = False) 
```
